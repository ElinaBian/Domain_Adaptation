```python
from data_process.usps_data import *
from data_process.mnist_data import *
from data_process.visualization import *
import matplotlib.pyplot as plt
import glob
from mcd_usps2mnist import *
```

```python
usps_images_train, usps_labels_train = load_usps_train_data()
usps_images_test, usps_labels_test = load_usps_test_data()

print("USPS_train_images shape: {}, USPS_train_labels shape: {}".format(usps_images_train.shape, usps_labels_train.shape))
print("USPS_test_images shape: {}, USPS_test_labels shape: {}".format(usps_images_test.shape, usps_labels_test.shape))
```

```python
ind_tr = np.random.choice(19999, 1800, replace = False)
ind_ts = np.random.choice(1500, 1200, replace = False)

usps_tr_im = np.reshape(usps_images_train, (-1, 784))
usps_tr_im = usps_tr_im[ind_tr,:]
usps_ts_im = np.reshape(usps_images_test, (-1, 784))
usps_ts_im = usps_ts_im[ind_ts,:]
usps_tr_labels = usps_labels_train
usps_tr_labels = usps_tr_labels[ind_tr,:]
usps_ts_labels = usps_labels_test
usps_ts_labels = usps_ts_labels[ind_ts,:]

ind_tr = np.random.choice(55000, 2000, replace = False)
ind_ts = np.random.choice(10000, 1500, replace = False)

mnist_images_train, mnist_labels_train, mnist_images_test, mnist_labels_test = load_mnist_data(28)
mnist_tr_im = mnist_images_train
mnist_tr_im = np.reshape(mnist_tr_im, (-1, 784))
mnist_tr_im = mnist_tr_im[ind_tr,:]
mnist_ts_im = mnist_images_test
mnist_ts_im = np.reshape(mnist_ts_im, (-1, 784))
mnist_ts_im = mnist_ts_im[ind_ts,:]
mnist_tr_labels = mnist_labels_train
mnist_tr_labels = mnist_tr_labels[ind_tr,:]
mnist_ts_labels = mnist_labels_test
mnist_ts_labels = mnist_ts_labels[ind_ts,:]
```

```python
filter_size = 5
n_filters_1 = 32
n_filters_2 = 48

n_units_g = 768   # 768 = 4*4*48
n_units_gc = 100
n_units_c = 100

learning_rate = 0.0002
epoch = 200
num_stepC = 2
batch_size = 64
show_step = 50
model_path = '/model'
is_saving = True
```

```python
mcd = MCD_DA()
mcd.train(images_train_1=usps_tr_im, labels_train_1=usps_tr_labels, images_test_1=usps_ts_im, \
          labels_test_1=usps_ts_labels, images_train_t=mnist_tr_im, labels_train_t=mnist_tr_labels,\
          images_test_t=mnist_ts_im, labels_test_t=mnist_ts_labels, \
          filter_size=filter_size, n_filters_1=n_filters_1, n_filters_2=n_filters_2, \
          n_units_g=n_units_g, n_units_c = n_units_c, n_units_gc = n_units_gc,\
          learning_rate = learning_rate, epoch = epoch, num_stepC = num_stepC, batch_size = batch_size, show_step = show_step, \
          is_saving = is_saving, model_path = model_path)
```
